```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame({
    'Square Foot':[1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000,1700, 2200, 2700, 3200, 3700, 4200, 4700, 5200, 5700, 6200],
    'Bedrooms':[2,3,4,4,3,5,4,5,6,7,2,3,4,4,3,5,4,5,6,7],
    'House Age':[10,15,20,25,30,35,40,45,50,55,12,18,22,28,32,38,42,48,52,60],
    'Price':[400000, 500000, 600000, 700000, 800000, 900000, 1000000, 1100000, 1200000, 320000, 420000, 520000, 620000, 720000, 820000, 920000, 1020000, 1120000, 1220000, 1300000]
})
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
x = df[['Square Foot','Bedrooms','House Age']]
y = df['Price']

model = LinearRegression().fit(x,y)
print(f"Intercept(c): {model.intercept_:.2f}")
print(f"coefficient(m): {model.coef_}")
```

```python
price = model.predict([[3500,3,30]])
print(f"\nPredicted Price: ${price[0]:.2f}")
```
