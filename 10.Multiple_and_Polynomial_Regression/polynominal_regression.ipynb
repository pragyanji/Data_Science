```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.DataFrame({
    'Temperature': [1, 2, 3, 4, 5,6,7,8],
    'Energy Consumption': [2,5,9,15,14,10,5,1]
})
```

```python
sns.scatterplot(data = df, x = 'Temperature', y = 'Energy Consumption')
plt.title("Energy Consumption vs Temperature")
```

```python
x = df[['Temperature']]
y = df['Energy Consumption']
x_poly = PolynomialFeatures(degree = 2).fit_transform(x)
print(x_poly)

model = LinearRegression().fit(x_poly, y)
energy_consumption_predicted = model.predict(PolynomialFeatures(degree = 2).fit_transform([[2]]))

print(f"Predicted Energy Consumption: {energy_consumption_predicted[0]:.2f}")

print(model.coef_)
print(model.intercept_)
```

```python
energy_consumption_pred = model.predict(x_poly)
print(energy_consumption_pred)
```

```python
# Using the model prediction.
sns.scatterplot(data=df, x='Temperature', y='Energy Consumption')
print(x_poly)
sns.lineplot(x=df['Temperature'], y=energy_consumption_pred, color='red')
plt.show()
```
