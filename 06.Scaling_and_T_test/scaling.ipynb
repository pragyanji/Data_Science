```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
# Create a simple series (eg, price in lakhs)
price = pd.Series([10, 20, 30, 40, 50],name='price')

#Scikit learn expects a 2D array, so we need to reshape/convert our 1D series to 2D array
# reshape(-1,1):
# -1 = 
# 1 = tells numpy to create exactly one column
prices_reshaped = price.values.reshape(-1,1)
prices_reshaped
```

```python
# Initialize scalers
std_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()


# create a comparison dataframe

results = pd.DataFrame({
    'Original': price,
    'Standardized (z-score)': std_scaler.fit_transform(prices_reshaped).flatten(),
    'Normalized (0-1)': minmax_scaler.fit_transform(prices_reshaped).flatten(),
    'RobustScale': robust_scaler.fit_transform(prices_reshaped).flatten()
})
results
```
