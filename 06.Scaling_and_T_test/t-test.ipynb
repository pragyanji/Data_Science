```python
import scipy as sp
import numpy as np
```

```python
# group A: sales without discount
group_a = [100, 110, 120, 130, 140]

# group B: sales with discount
group_b = [150, 170, 160, 190, 250]

# Perform t-test
t_statistic, p_value = sp.stats.ttest_ind(group_a, group_b)

print(f"P-value: {p_value:.4f}")
print(f"T-statistic: {t_statistic:.4f}")

if p_value < 0.05:
    print("The discount has a statistically significant effect on sales.")
else:
    print("The difference might just be luck, and the discount may not have a significant effect on sales.")

# explanation of p_value
# The p-value represents the probability of observing the data (or something more extreme) assuming that the null hypothesis is true. In this case, the null hypothesis is that there is no difference in sales between the two groups (with and without discount). A low p-value (typically less than 0.05) indicates that we can reject the null hypothesis, suggesting that there is a statistically significant difference in sales due to the discount. Conversely, a high p-value suggests that we fail to reject the null hypothesis, meaning that any observed difference in sales could be due to random chance rather than the effect of the discount.
```
