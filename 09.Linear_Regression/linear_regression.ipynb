```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    'X': [1400,1600,1700,1875,1100,1550,2350,2450,1425,1700],
    'Y': [245,312,279,308,199,219,405,324,319,255]
})
sns.scatterplot(x='X',y='Y',data=data)
plt.show()
```

```python
model = LinearRegression().fit(data[['X']], data[['Y']])
print(f"Intercept(c): {model.intercept_[0]:.2f}")
print(f"Coefficient(m): {model.coef_[0][0]:.2f}")
```

```python
price = model.predict([[1800]])
print(f"Predicted price for 1800 sq ft: {price[0][0]:.2f}")
```

```python
sns.scatterplot(data=data,x='X',y='Y')
y_all_predicted = model.predict(data[['X']])
sns.lineplot(x=data['X'],y=y_all_predicted.flatten(),color='red')
plt.show()
```
