```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
df
```

```python
print(f"{df['Hours_Studied'].isnull().sum()} null values in 'Hours_Studied'")
print(f"{df['Exam_Score'].isnull().sum()} null values in 'Exam_Score'")
```

```python
# Scatter plot using Matplotlib of 'Hours_Studied' vs 'Exam_Score'
plt.figure(figsize=(10, 6))
plt.scatter(df['Hours_Studied'], df['Exam_Score'], color='blue', alpha=0.3)
plt.title('Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```
